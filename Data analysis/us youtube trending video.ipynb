```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset
data = pd.read_csv("USvideos.csv")  # or another country file
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
data['publish_time'] = pd.to_datetime(data['publish_time'])
data.isnull().sum()
```

```python
data['publish_date'] = data['publish_time'].dt.date
data['publish_hour'] = data['publish_time'].dt.hour
```

```python
category_stats = data.groupby('category_id')['views'].sum().sort_values(ascending=False)
px.bar(category_stats, title='Total Views by Category')
```

```python
top_channels = data.groupby('channel_title')['views'].sum().sort_values(ascending=False).head(10)
px.bar(top_channels, title='Top 10 Channels by Views')
```

```python
sns.heatmap(data[['views', 'likes', 'comment_count']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Engagement Metrics')
plt.show()
```

```python
hourly_views = data.groupby('publish_hour')['views'].mean()
px.line(hourly_views, title='Average Views by Publish Hour')
```

```python
data['like_dislike_ratio'] = data['likes'] / (data['dislikes'] + 1)
sns.boxplot(x='category_id', y='like_dislike_ratio', data=data)
```
